### Draw Down Proxy
<b>Draw Down Definition: </b>how much your profit drops from a peak before recovering.<br>
<p>Since the <b>balance</b> is not mentioned in the dataset, we'll try to estimate the Draw Down via Average Loss and Max Loss, Total number
    of losses per each group as a proxy.</p>
<p>
    While Greed days appear more profitable on average, they carry significantly higher tail risk. The average loss in Greed (-$164$) is lower than in Fear (-$171$), suggesting better routine discipline. However, the drawdown proxy (Max Single Hit) is 230% higher in Greed. This indicates that Greed environments foster 'Black Swan' events where a trader fails to manage a catastrophic loss, potentially wiping out weeks of accumulated profit in a single execution.
</p>

<h3>High Frequency Scalper</h3>
<p>
    The Scalper (High Frequency): High trade count ($11767$(median) trades), high Win Rate ($>80\%$), but low Average P&L ($<$41$).
        <br>Evidence: These traders likely thrive in Fear days, grinding out small gains during volatility.
</p>
        
<h3>Aggressive Gambler</h3>
<p>
    The Gambler (High Leverage): High $\sigma$ and a Max Single Hit that is an outlier (like the account that hit the $117k$ loss).
    <br>
    Behavior: They show "explosive" growth during Greed but are the primary cause of your group's massive drawdowns due their outlier nature.
</p>
<h3>Standard Swingger</h3>
<p>
    The Swing Trader (Low Frequency): Low trade count ($<500$ trades), moderate Win Rate ( $50 $ to $60$), but high Average P&L ($200$).
        <br>Evidence: These traders likely dominate the Greed days, holding positions longer to capture bigger moves.
</p>
<h3>Standard Retail Trader</h3>
<p>The retail trader is the one that doesnt fit into neither of these categories</p>

```python

def identify_segment(row):
    if row['Total_Trades'] > 11767 and row['Win_Rate'] > 80 and row.Avg_PnL < 41:
        return 'High-Freq Scalper'
    elif row['Max_Hit'] < -24126:
        return 'Aggressive Gambler'
    elif row['Avg_PnL'] > 100:
        return 'Strategic Swing'
    else:
        return 'Standard Retail'
dashboard['Segment'] = dashboard.apply(identify_segment, axis=1)
```
